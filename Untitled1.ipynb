```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
x=sns.load_dataset('tips')
print(x)
```

```python
sns.displot(x['total_bill'],bins=[10,20,30,40,50],kde=True)
plt.show()
```

boxplot

```python
p=[10,20,30,40,50]
h=[15,25,35,45,55]
t=[p,h]
plt.boxplot(t,labels=['python','java'])    #boxplot is a graph of matplotlib
plt.show()
```

multiple col in displot

```python
sns.displot(x[['total_bill','smoker']],x='total_bill',multiple='stack',col='smoker')
plt.show()
```

```python
iphone=[9,7,20,12,34]
samsung=[23,43,54,56,76]
pixel=[8,5,23,54,56]
details=['camera','display','storage','battery','price']
a=np.linspace(0,2*np.pi,len(details))
plt.figure(figsize=(5,5))
plt.subplot(111,polar=True)
plt.plot(a,iphone,label='iphone')
plt.plot(a,samsung,label='samsung')
plt.plot(a,pixel,label='pixel')
plt.thetagrids(a*180/np.pi,details)
# plt.title(iphone')
plt.show()
```

```python
plt.polar(iphone)
plt.thetagrids(a*180/np.pi,details)
plt.show()
```

# **Matplotlib(library)**


```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,20,30,40,50 ]

plt.plot(x,y,marker='^',c='r',mfc='y',mec='y',ls='dotted',ms=15,lw=1,label='simple')
plt.legend(loc='lower right',facecolor='r',framealpha=1,shadow=True,fancybox=True,edgecolor='b')
plt.xticks(x)
plt.yticks(y)
plt.xlabel('x-axis',fontsize=20)
plt.ylabel('y-axis')
plt.title('Graph')
plt.grid()
plt.show()
```

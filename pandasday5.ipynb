# **PANDAS DATE AND TIME FUNCTION**

*   1)HOUR
*  2)MINUTE
*  3)DAY
*   4)YEAR
*   5)MONTH
*   6)WEEKDAY







```python
import pandas as pd
date_time=['2025-03-06 10:02','2024-07-11 11:45','2023-08-20 09:30']
x={'date_time':date_time}
df=pd.DataFrame(x)
df['date_time']=pd.to_datetime(df['date_time'])     #to_datetime is a function (converting the dataframe to date_time )
df['hour']=df['date_time'].dt.hour                  #here we use the dt.hour function
df
df['minute']=df['date_time'].dt.minute              #here we have used the dt.minute function
df
df['year']=df.date_time.dt.year
df
df['day0fweek']=df.date_time.dt.day_name()          #separate column(   var_name['col_name']=var_name.array_name/var_name['array_name']    )
df
df['month']=df.date_time.dt.month_name()
df
df['weekday']=df.date_time.dt.weekday
df
# df['age']=df.date_time
# df
# df['bro']=df['date_time']                     # adding column to the date_time table will add same dat_time table with the specified number
# df
```

```python
import pandas as pd
data= {'EID':[89,23,50,100,300],
      'ESAL':[1000,2000,3000,4000,5000]}
df=pd.DataFrame(data)
df.sort_index()  # by default ascending
df.sort_index(ascending=False) # descending order


```

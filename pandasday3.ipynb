# **iloc**

```python
import pandas as pd

# Creating Employee DataFrame
data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Department": ["HR", "IT", "Finance", "Marketing", "Sales"],
    "Salary": [50000, 60000, 55000, 52000, 58000],
    "Joining_Date": ["2020-06-15", "2019-08-01", "2021-01-10", "2018-07-20", "2022-03-05"]
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)

'''dataframe.iloc[row-number,col-number]'''
print(df.iloc[1,3])           #display the row data from particular column

'''dataframe.iloc[row-start:row-end,col-start:col-end]'''
print(df.iloc[1:3,0:2])       #display the data from start row and stop row as well as col start and col end

'''dataframe.iloc[start:stop,col-number]'''
print(df.iloc[0:3,2])         #display the data from the column upto the end index

'''dataframe.iloc[row-number1,row-number2,.......row-numbern]'''
print(df.iloc[[0,4,2]])       #use to display the multiple rows

```

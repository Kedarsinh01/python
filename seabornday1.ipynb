```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

m=sns.load_dataset('penguins')
print(m)
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=m,hue='species',style='species',legend=True,markers=["*","d","o"],palette=['r','b','g'])
plt.show()
```

```python
sns.barplot(x='species',y='bill_depth_mm',data=m,hue='species',errorbar=('ci',100),err_kws={'color':'red'},palette=['r','b','g'],capsize=.3)
plt.show()
```

```python
d=sns.load_dataset('tips').head(30)
print(d)
```

```python
sns.lineplot(x='total_bill',y='tip',data=d,hue='sex',legend=True,style='sex',markers=["*","d"],ms=10,mec='black',dashes=False)
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=d,hue='sex',style='sex',legend=True,markers=["*","d"],palette=['g','b'])
plt.show()
```

```python
sns.barplot(x='total_bill',y='tip',data=d,hue='sex',palette=['r','b','g'])
plt.show()
```
